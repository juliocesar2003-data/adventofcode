```python
with open('D:/Pythonpipi/adventfive.txt', 'r') as file:
    content = file.read()
parts = content.strip().split("\n\n")
a = parts[0].splitlines()
d = parts[1].splitlines()
```

```python
cont = 0
sts = []

for b in a:
    temp = b.find("-")
    sts.append((int(b[:temp]),int(b[temp +1:])))
```

```python
for e in d:
    for h in sts:
        if int(e) >= h[0] and int(e) <= h[1]:
            cont += 1 
            break
print(cont)
```

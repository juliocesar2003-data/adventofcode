```python
with open('D:/Pythonpipi/adventfive.txt', 'r') as file:
    content = file.read()
parts = content.strip().split("\n\n")
a = parts[0].splitlines()
```

```python
soma = 0
lis = []
sil = []

for b in a:
    temp = b.find("-")
    temp2 = (int(b[:temp]),int(b[temp +1:]))
    lis.append(temp2)

lis = sorted(lis)
```

```python
inicio = lis[0][0]
fim = lis[0][1]

for c in range(1,len(lis)):
    v2ini = lis[c][0]
    v2fim = lis[c][1]
    if v2ini <= fim:
        if v2fim > fim:
            fim = v2fim
    else:
        sil.append((inicio, fim))
        inicio = v2ini
        fim = v2fim
sil.append((inicio, fim))
```

```python
for d,e in sil:
    soma += (e - d + 1)

print(soma)
```

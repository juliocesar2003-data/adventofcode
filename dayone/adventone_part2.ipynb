```python
with open('adventone.txt', 'r') as file:
    lines = file.read().splitlines() 
```

```python
start = 50  
cont2 = 0 
```

```python
for item in lines: #acessa cada valor da lista
    letter = item[0] # o primeiro valor é sempre a direção L ou R
    number = int(item[1:]) # o restante do valor é a distância
    if letter == 'L':
        number *= -1 # Vou considerar os valores de L como negativos, já que giram o cofre no sentido anti-horário
    temp0 = start #guarda o valor do ponteiro
    start += number # calcula onde vai parar o ponteiro
    if letter == 'L': # se a direção for L teste todos os valores para checar se é 0
        for j in range(temp0-1,start-1,-1): 
            if j % 100 == 0: # valores como 100, 200 e 300 também são considerados
                cont2 += 1
    else: #mesmo cálculo, mas agora para R
        for j in range(temp0+1,start+1,1):
            if j % 100 == 0:
                cont2 += 1
    if start >= 100 or start <= 100: # se o valor sair da faixa '0-99', encaixa ele no valor correspondente no cofre
        temp = int(start / 100)
        start = start - 100 * temp 
    elif start <0:
        start += 100 
```

```python
print(f'Password2: {cont2}')
```

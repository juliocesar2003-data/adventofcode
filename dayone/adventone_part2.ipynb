```python
with open('adventone.txt', 'r') as file:
    lines = file.read().splitlines() 
```

```python
start = 50  
cont2 = 0 
```

```python
for item in lines: 
    letter = item[0] 
    number = int(item[1:]) 
    if letter == 'L':
        number *= -1 
    temp0 = start 
    start += number 
    if letter == 'L': 
        for j in range(temp0-1,start-1,-1): 
            if j % 100 == 0: 
                cont2 += 1
    else: 
        for j in range(temp0+1,start+1,1):
            if j % 100 == 0:
                cont2 += 1
    if start >= 100 or start <= 100: 
        temp = int(start / 100)
        start = start - 100 * temp 
    elif start <0:
        start += 100 
```

```python
print(f'Password2: {cont2}')
```

```python
with open('adventone.txt', 'r') as file:
    lines = file.read().splitlines() 
```

```python
start = 50  
count = 0 
```

```python
for item in lines: 
    letter = item[0] 
    number = int(item[1:]) 
    if letter == 'L':
        number *= -1 
    start += number 
    if start >= 100 or start <= 100: 
        temp = int(start / 100)
        start = start - 100 * temp 
    elif start <0:
        start += 100 

    if start == 0:
        count +=1
```

```python
print(f'Password1: {count}')
```

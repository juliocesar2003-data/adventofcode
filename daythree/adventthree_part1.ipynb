```python
with open('D:/Pythonpipi/adventthree.txt', 'r') as file:
    raw_text = file.read()
a = raw_text.split()
```

```python
soma = 0
for i in a:
    temp = i[:-1]
    valorp1 = max(temp)
    localp1 = i.index(valorp1)
    valorp2 = max(i[localp1 + 1:])
    voltage = valorp1 + '' + valorp2
    soma += int(voltage)
print(soma)
```

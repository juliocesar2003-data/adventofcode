```python
with open('D:/Pythonpipi/adventthree.txt', 'r') as file:
    raw_text = file.read()
a = raw_text.split()
```

```python
soma = 0
for i in a:
    lop = 12 
    voltage = ''
    temp = 0
    for j in range(lop-1,-1,-1):
        if j > 0:
            j *= -1
        else:
            j = None   
        tempo = i[temp:j]
        valor = max(tempo)
        temp += tempo.index(valor) + 1
        voltage += valor
    soma += int(voltage)

print(soma)
```

```python
with open('D:/Pythonpipi/adventfour.txt', 'r') as file:
    raw_text = file.read()
a = raw_text.split()
```

```python
u = []
for k in a:
    u.append(list(k))
a = u
cont2 = 0
flag = True
b = a
```

```python
while flag:
    a = b
    temp = cont2
    for i in range(0,len(a)):
        for j in range(0,len(a[i])):
            if a[i][j] == "@":
                cont = 0
                for h in (i-1,i,i+1):
                    if h < 0 or h == len(a):
                        pass
                    else:
                        for g in (j-1,j,j+1):
                            if (h == i and g == j) or g < 0 or g == len(a[i]):
                                pass
                            else:
                                if a[h][g] == "@":
                                    cont += 1       
                if cont < 4:
                    cont2 += 1
                    b[i][j] = "."
            else:
                pass
    if temp == cont2:
            flag = False
print(cont2)
```

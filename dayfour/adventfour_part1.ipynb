```python
with open('adventfour.txt', 'r') as file:
    raw_text = file.read()
a = raw_text.split()
```

```python
cont = 0
cont2 = 0

for i in range(0,len(a)):
    for j in range(0,len(a[i])):
        if a[i][j] == "@":
            cont = 0
            for h in (i-1,i,i+1):
                if h < 0 or h == len(a):
                    pass
                else:
                    for g in (j-1,j,j+1):
                        if (h == i and g == j) or g < 0 or g == len(a[i]):
                            pass
                        else:
                            if a[h][g] == "@":
                                cont += 1       
            if cont < 4:
                cont2 += 1
        else:
            pass
print(cont2)         
```

```python
with open('D:/Pythonpipi/adventtwo.txt', 'r') as file:
    raw_text = file.read().strip()
a = raw_text.split(',')
```

```python
cont = 0
```

```python
# couldn`t do a loop that works for every single number. So ran a max(len()) (10) and make the code for that
for b in a:
	temp = b.find("-")
	for c in range(int(b[:temp]),int(b[temp +1:])+1):
		c = str(c)
		siz = len(c)
		if siz in (2,4,8):
			indx = siz // 2
			if c[:indx] == c[indx:]:
				cont += int(c)
		elif siz == 6:
			if c[:3] == c[3:]:
				cont += int(c)
			elif c[:2] == c[2:4] == c[4:]:
				cont += int(c)
		elif siz == 10:
			if c[:5] == c[5:]:
				cont += int(c)
			elif c[:2] == c[2:4] == c[4:6] == c[6:8] == c[8:]:
				cont += int(c)			
		elif siz == 9:
			if c[:3] == c[3:6] == c[6:]:
				cont += int(c)
		elif siz == 1:
			pass
		else:
			if c.count(c[0]) == siz:
				cont += int(c)
```

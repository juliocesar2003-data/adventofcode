```python
with open('adventtwo.txt', 'r') as file:
    raw_text = file.read().strip()
a = raw_text.split(',')
```

```python
cont = 0
```

```python
for b in a:
	temp = b.find("-")
	for c in range(int(b[:temp]),int(b[temp +1:])+1):
		c = str(c)
		siz = len(c)
		if siz % 2 == 0:
			indx = siz // 2
			if c[:indx] == c[indx:]:
				cont += int(c)
```

```python
print(cont)
```
